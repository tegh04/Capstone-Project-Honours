### Create animation from the files

```python
img.shape
```

```python
# The GIF cannot be created unless the data type is np.uint8 for an RGB image
imgs = [Image.fromarray(frame) for frame in (img[:, :, :, 3]/65535*255).astype(np.uint8)]

# duration is the number of milliseconds between frames; this is 40 frames per second
imgs[0].save("image set1.gif", save_all=True, append_images=imgs[1:], duration=80, loop=0)
```
